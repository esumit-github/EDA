### Replacing categorical features

```python
# finding the categorical features

categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes=='O']

categorical_features
```

- 1. Technique for missing values in the dataset
    - to make it uniform rplace all categorical features which have NA values with 'Missing'
    - The function replace_categorical_features dose this work.

```python
data = dataset.copy()
def replace_categorical_features(dataset, categorical_features):
    data[categorical_features]= data[categorical_features].fillna('Missing')
    return(data)
```

```python
replace_categorical_features(data, categorical_features)
```

- 2. This Technique is to replace the missing values with the mode of that feature value

```python
data = dataset.copy()
def replace_categorical_features(dataset, categorical_features):
    data[categorical_features]= data[categorical_features].fillna(data[categorical_features].mode())
    return(data)
```

```python
replace_categorical_features(data, categorical_features)
```

The 2nd technique will not be valid as Features are having NaN values after this replacemnt

```python
# Check the Numerical values and NaN

numerical_features_NaN = [feature for feature in dataset.columns if dataset[feature].dtypes!='O' and
                          dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_NaN
```

```python
for feature in numerical_features_NaN:
    ## as there are outliers in the data set replace them with median values
    median_value = dataset[feature].median
    
    ## create a new feature called "feature+nan" to capture nan values of that 
    # feature. Creating a new column named "feature+nan"
    ## in the data set the features which have the nan values replace 
    # them with the 1 else with 0
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    
    # replase the feature with the median values
    dataset[feature].fillna(median_value, inplace=True)
```
