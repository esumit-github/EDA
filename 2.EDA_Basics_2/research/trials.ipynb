# Handling missing data

```python
# check how many null values are there accross the features
dataset.isnull().sum()
```

```python
null_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]

# Alteranative
#null_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]

# Advanced code to fins the percentage of null values in the dataset per feature
for feature in null_feature:
    print("The Feature is {} and the missing value is: {}%".format(feature, np.round(dataset[feature].isnull().mean(),4)))
```
